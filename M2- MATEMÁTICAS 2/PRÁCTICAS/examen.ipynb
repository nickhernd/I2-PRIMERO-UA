```python
#ejercicio 1
matrix = [[1,2,'a'],[3,'z',16]]
solo_numeros = []

for fila in matrix:
    for elemento in fila:
        if isinstance(elemento, int):
            solo_numeros.append(elemento)

print(solo_numeros)

pares = [i*i for i in solo_numeros if(i%2==0)]
print(pares)
```

```python
#ejercicio 2
from sympy import *
from sympy.abc import x
from sympy.plotting import plot

f1 = x**4-6*x**2+24*x+16

plot(f1, (x,-10,10))

criticosX = list(solveset(f1,x))
print(criticosX)
criticosY = f1.subs(x,0)
print(criticosY)

dfx = diff(f1,x)
dfxx = diff(dfx, x)

puntos_crit = list(solveset(dfx,x))
print(puntos_crit)


        
```
